```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
E0=10
R=1
D=7000
l=0.0006
k=2*np.pi/l

Rspace=np.linspace(0,R,51)
phi=np.linspace(0,2*np.pi,101)
Rspace,phi=np.meshgrid(Rspace,phi)

x=np.linspace(-10,10,201)
y=np.linspace(-10,10,201)

E=np.zeros((len(x),len(y)),dtype=complex)
for i in range(len(x)):
    for j in range(len(y)):
        phase=k*(Rspace*np.cos(phi)*x[i]/D+Rspace*np.sin(phi)*y[j]/D)
        dE=np.exp(phase*1j)
        E[i,j]=sum(sum(dE))
```

```python
def sinq(x,N):
    if x!=0:
        return np.sin(N*x)/np.sin(x)
    else: return N
d=2
N=4
x=np.linspace(-15,15,201)    
I=(E0/D)**2*(1/2)*np.real(E*np.conj(E))*(np.sin(N*k*(x/D)*d/2)/np.sin(k*(x/D)*d/2))**2

plt.figure(figsize=[6,6])
plt.pcolormesh(x,y,I,vmin=0,vmax=N**2,cmap='gray')
plt.xlabel('x [mm]')
plt.ylabel('y [mm]')
plt.show()
```

```python
#%%
x=np.linspace(-30,30,1001)
y=np.linspace(-30,30,1001)
x,y=np.meshgrid(x,y)
N=4
theta=np.linspace(0,np.pi,N)
r=1
a=r*np.sin(theta)
d=r*np.cos(theta)
b=0.5
alphax=k*(b/2)*x/D
I0=10

sum_b=np.sinc(alphax)

sum_a=0
for p in range(len(a)):
    alphay=k*(a[p]/2)*(y/D)
    sum_a=sum_a+np.sinc(alphay)*np.exp(1j*k*(x/D)*d[p])
    #print(sum_a)
sumsum=(1/2)*np.real(sum_a*np.conj(sum_a))
#print(sumsum)
Ir=I0*sumsum*sum_b**2

plt.figure(figsize=[6,6])
plt.pcolormesh(x,y,Ir,vmin=0,vmax=N^2,cmap='gray')
plt.xlabel('x [mm]')
plt.ylabel('y [mm]')
plt.show()
```
